# 결손 데이터 처리하기

- 결손 데이터 여부 확인 : isna()

````python
titanic_df.isna().head(3)
````

- 결손 데이터 대체하기 : fillna()

````python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
````

- 'Age'컬럼의 NaN값을 평균나이로 'Embarked'컬럼은 NaN값을 'S'로 대체

````python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
````

# apply lambda식으로 데이터 가공

````python
def get_square(a):
    return a**2

print("3의 제곱은:", get_square(3))
````

````python
lambda_square = lambda a: a**2
print("3의 제곱은:", lambda_square(3))
````

````python
a = [1,2,3]
squares = map(lambda x: x**2, a)
list(squares)
````

- 'Name'컬럼의 문자열 개수를 별도의 컬럼은 'Name_len'으로 생성

````python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head(3)
````

- 나이가 15세 미만이면 'Child' 그렇지 않으면 'Adult'를 새로운 컬럼 'Child_Adult'에 삽입

````python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x:'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
````

- lambda 식은 if else를 지원할 때 if절의 경우 if식보다 반환 값이 먼저 기술되어야 

    ex) if x <== 15: 'Child' else 'Adult'  --> 'Child' if x <= 15 else 'Adult'

- 나이가 15세 이하이면 Child, 15 ~ 60세 사이는 Adult, 61세 이상은 Elderly로 분류하여 새로운 컬럼 Age_Cat 삽입

````python
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x : 'Child' 
                                                if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_Cat'].value_counts()
````

````python
def get_category(age):
    cat = ''
    if age <=5: cat = 'baby'
    elif age <=12: cat = 'child'
    elif age <=18: cat = 'teenager'
    elif age <=25: cat = 'student'
    elif age <=35: cat = 'young adult'
    elif age <=60: cat = 'adult'
    else: cat = 'Elderly'
    
    return cat

titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_Cat']].head(3)
````
