```python
# First Machine Learning - predict iris
- sklearn.datasets : 사이킷런 자체적으로 제공하는 데이터 세트 생성 모듈
- sklearn.tree : 트리 기반 ML 알고리즘 모듈
- sklearn.model_selection : 학습데이터와 검증데이터, 예측데이터 분리하거나 최적의 하이퍼파라미터 평기 모듈
- data load : load_iris()이용
- ML 알고리즘 : 의사결정트리 (Decision Tree) - DecisionTreeClassifier
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

# 데이터 세트 로딩
iris = load_iris()
iris_data = iris.data

# iris.target은 붓꽃 데이터 세트에서 레이블(결정값) 데이터를 numpy로 가지고 있음
iris_label = iris.target
print("target값:", iris_label)
print("target명:", iris.target_names)
```

```python
iris
```

```python
# 붓꽃 데이터 세틀를 자세히 보기 위해 DataFrame으로 변환
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label'] = iris.target
iris_df.head(3)
```

```python
# 데이터 분리 :학습용/테스트용
# train_test_split()
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
```

# 학습과 예측 : DecisionTreeClassifier

```python
#  DecisionTreeClassifier 객체 생서
dt_clf = DecisionTreeClassifier(random_state=11)

# 학습숫행
dt_clf.fit(X_train, y_train)
```

```python
# 예측
pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("예측 정확도:{0: .4f}".format(accuracy_score(y_test, pred)))
```
